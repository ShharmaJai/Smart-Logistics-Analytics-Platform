```python
# Insert warehouses
for index, row in df_warehouses.iterrows():
    cursor.execute("INSERT IGNORE INTO warehouses (warehouse_id, city, state, capacity) VALUES (%s, %s, %s, %s)",
    (row['warehouse_id'], row['city'], row['state'], row['capacity']))
conn.commit()
print("Warehouses done!")

# Insert courier_staff
for index, row in df_courier_staff.iterrows():
    cursor.execute("INSERT IGNORE INTO courier_staff (courier_id, name, rating, vehicle_type) VALUES (%s, %s, %s, %s)",
    (row['courier_id'], row['name'], row['rating'], row['vehicle_type']))
conn.commit()
print("Courier staff done!")

# Insert routes
for index, row in df_routes.iterrows():
    cursor.execute("INSERT IGNORE INTO routes (route_id, origin, destination, distance_km, avg_time_hours) VALUES (%s, %s, %s, %s, %s)",
    (row['route_id'], row['origin'], row['destination'], row['distance_km'], row['avg_time_hours']))
conn.commit()
print("Routes done!")

# Insert shipments
for index, row in df_shipments.iterrows():
    cursor.execute("INSERT IGNORE INTO shipments (shipment_id, order_date, origin, destination, weight, courier_id, status, delivery_date) VALUES (%s, %s, %s, %s, %s, %s, %s, %s)",
    (row['shipment_id'], row['order_date'], row['origin'], row['destination'], row['weight'], row['courier_id'], row['status'], row['delivery_date']))
conn.commit()
print("Shipments done!")

# Insert shipment_tracking
for index, row in df_shipment_tracking.iterrows():
    cursor.execute("INSERT IGNORE INTO shipment_tracking (tracking_id, shipment_id, status, timestamp) VALUES (%s, %s, %s, %s)",
    (row['tracking_id'], row['shipment_id'], row['status'], row['timestamp']))
conn.commit()
print("Tracking done!")

# Insert costs
for index, row in df_costs.iterrows():
    cursor.execute("INSERT IGNORE INTO costs (shipment_id, fuel_cost, labor_cost, misc_cost) VALUES (%s, %s, %s, %s)",
    (row['shipment_id'], row['fuel_cost'], row['labor_cost'], row['misc_cost']))
conn.commit()
print("Costs done!")

print("All data loaded into MySQL!")
```
